```python
# 1. Overview of the Data
import pandas as pd
data=pd.read_csv(r"C:\Users\PERSONAL\Downloads\country_wise_latest.csv")
print("Data Overview:")
print(data.info())  # Shows data types and missing values
print("\nFirst 5 Rows of the Data:")
print(data.head())

# 2. Check for Missing Data
print("\nMissing Values in Each Column:")
print(data.isnull().sum())

# 3. Basic Descriptive Statistics
print("\nDescriptive Statistics of Key Columns:")
print(data[['Confirmed', 'Deaths', 'Recovered', 'Active', 'New cases', 'New deaths']].describe())

# 4. Top 5 Countries by Confirmed Cases
top_countries_confirmed = data[['Country/Region', 'Confirmed']].sort_values(by='Confirmed', ascending=False).head(5)
print("\nTop 5 Countries with the Highest Confirmed Cases:")
print(top_countries_confirmed)

# 5. Case Fatality Rate (CFR) - Deaths/Confirmed * 100
data['CFR'] = (data['Deaths'] / data['Confirmed']) * 100
top_cfr = data[['Country/Region', 'CFR']].sort_values(by='CFR', ascending=False).head(5)
print("\nTop 5 Countries by Case Fatality Rate (CFR):")
print(top_cfr)

# 6. Countries with the Highest Percentage Increase in Cases in the Last Week
data['1 week % increase'] = pd.to_numeric(data['1 week % increase'], errors='coerce')  # Ensure numeric values
top_week_increase = data[['Country/Region', '1 week % increase']].sort_values(by='1 week % increase', ascending=False).head(5)
print("\nTop 5 Countries with the Highest Percentage Increase in Cases in the Last Week:")
print(top_week_increase)
```
